```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('graphdesign.mplstyle')

import my_functions as mf
```

```python
# demo sigmnal
print("Demo signal")
N_t = 256            # サンプル数
dt = 0.01          # サンプリング間隔
f1, f2 = 10, 30    # 周波数
t = np.arange(0, N_t*dt, dt)
f = 8*np.sin(2*np.pi*f1*t) + 4*np.sin(2*np.pi*f2*t) + 0.3 * np.random.randn(N_t)
   
max_f = 50
```

```python
mf.plot_fft(t, f, max_f, "img_fft.png")
```

```python
f_hpf = mf.FT_Filter(f, t, "hpf", 20)
mf.plot_fft(t, f_hpf, max_f, "img_HPF.png")
```

```python
f_lpf = mf.FT_Filter(f, t, "lpf", 20)
mf.plot_fft(t, f_lpf, max_f, "img_LPF.png")
```
